# 2m Air and Soil Temperature Analysis

## Load Data

The test dataset contains 2m Air Temperature and Soil Temperature Level 1 at 12:00 for every day in April. Years covered: 1950 - 2025

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Plot style for the notebook
plt.style.use('seaborn-v0_8-whitegrid')

# 1. Load the dataset
file_path = r"C:\Users\James\Downloads\2msoil-allyears.nc" 
data = xr.open_dataset(file_path)

# Check data loaded correctly
print("Dataset loaded successfully.")
print("Dimensions:", data.dims)
```

## Computation

```python
# Define linear function
def linear_fit(x, m, c):
    return m * x + c

# Configuration
variables = ['t2m', 'stl1']
labels = ['2m Air Temp', 'Soil Temp (Lvl 1)']
colors = ['blue', 'green']
results = {}  # Dictionary to store calculation results for plotting

print("--- Analysis Results ---")

for var_name, label, color in zip(variables, labels, colors):
    
    # 1. Average over spatial dimensions
    # Collapses lat/lon, leaving one value per time step
    time_series = data[var_name].mean(dim=['latitude', 'longitude'])
    
    # 2. Extract X (Years) and Y (Temperature)
    # Using 'valid_time' coordinate from your dataset
    x_years = time_series['valid_time'].dt.year.values
    y_temps = time_series.values
    
    # 3. Fit the data
    popt, pcov = curve_fit(linear_fit, x_years, y_temps)
    
    # Calculate uncertainties
    m_fit, c_fit = popt
    m_err, c_err = np.sqrt(np.diag(pcov))
    
    # Store results for the next cell
    results[label] = {
        'x': x_years,
        'y': y_temps,
        'popt': popt,
        'color': color,
        'slope': m_fit,
        'slope_err': m_err
    }
    
    print(f"\n{label}:")
    print(f"  Slope: {m_fit:.4f} ± {m_err:.4f} K/year")
    print(f"  Intercept: {c_fit:.4f} ± {c_err:.4f} K")
```

## Visualisation

```python
plt.figure(figsize=(10, 6))

for label, res in results.items():
    x = res['x']
    y = res['y']
    popt = res['popt']
    color = res['color']
    
    # Plot raw data points
    plt.plot(x, y, 'o', color=color, alpha=0.5, label=f'{label} (Data)')
    
    # Plot fitted line
    plt.plot(x, linear_fit(x, *popt), '-', color=color, linewidth=2, label=f'{label} (Fit)')

# Formatting
plt.title('UK Temperature Trends: 2m Air vs Soil (Level 1)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Temperature (K)', fontsize=12)
plt.legend(frameon=True, fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## Extending to Yearly Average

Now considering all months

```python
# --- EXTENSION: Annual Averaging from Monthly Data ---

# 1. Load the new dataset (all months)
new_file_path = r"C:\Users\James\Downloads\2msoil-allyearsfull.nc"

data_all_months = xr.open_dataset(new_file_path)

# Check if expver
if 'expver' in data_all_months.coords:
    try:
        data_all_months = data_all_months.combine_first(data_all_months.sel(expver=1))
    except:
        pass

plt.figure(figsize=(10, 6))

# Loop through variables (uses names defined in previous cells)
for var_name, color, label in zip(variables, colors, labels):
    
    # --- STEP 1: Spatial Average ---
    # Collapse lat/lon to get a time series (1 point per month)
    monthly_series = data_all_months[var_name].mean(dim=['latitude', 'longitude'])
    
    # --- STEP 2: Temporal Average (Resample) ---
    # '1Y' tells xarray to group data by 1 Year and take the mean
    # This turns 12 monthly points into 1 annual point
    annual_series = monthly_series.resample(valid_time='1Y').mean()
    
    # Prepare data for plotting/fitting
    x_years = annual_series['valid_time'].dt.year.values
    y_temps = annual_series.values
    
    # --- STEP 3: Fit and Plot ---
    popt, pcov = curve_fit(linear_fit, x_years, y_temps)
    m_fit, c_fit = popt
    m_err, c_err = np.sqrt(np.diag(pcov))
    
    print(f"--- Annual Analysis for {label} ---")
    print(f"Slope: {m_fit:.4f} ± {m_err:.4f} K/year")
    
    # Plot
    plt.plot(x_years, y_temps, 'o', color=color, alpha=0.5, label=f'{label} (Annual Mean)')
    plt.plot(x_years, linear_fit(x_years, *popt), '-', color=color, linewidth=2)

plt.title('UK Annual Mean Temperature Trends (Averaged over 12 Months)')
plt.xlabel('Year')
plt.ylabel('Temperature (K)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
