# Comparison of Classifiers (Z-score)

```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GroupShuffleSplit # <--- NEW IMPORT
from sklearn.metrics import accuracy_score, roc_curve, auc

# --- 1. DATA PREPARATION ---
FILE_PATH = r'C:\Users\James\Downloads\era5new.nc'
MEAN_VARS = ['t2m', 'stl2', 'wind_speed', 'd2m', 'swvl1', 'sp']
SUM_VARS = ['tp', 'e', 'ro', 'sf']
VAR_LIST = MEAN_VARS + SUM_VARS

def load_and_prep_seasonal_data(file_path=FILE_PATH):
    print("Loading and structuring seasonal dataset...")
    ds = xr.open_dataset(file_path)
    if 'valid_time' in ds.coords: ds = ds.rename({'valid_time': 'time'})
        
    u_var, v_var = ds.get('u10', ds.get('10u')), ds.get('v10', ds.get('10v'))
    if u_var is not None and v_var is not None: ds['wind_speed'] = np.sqrt(u_var**2 + v_var**2)
    if 'sp' in ds: ds['sp'] = ds['sp'] / 100.0
    for var in ['t2m', 'stl2', 'd2m']:
        if var in ds and ds[var].mean() > 200: ds[var] = ds[var] - 273.15
            
    available_vars = list(set(VAR_LIST).intersection(ds.data_vars))
    df = ds[available_vars].to_dataframe().dropna().reset_index()
    
    df['month'], df['year'] = df['time'].dt.month, df['time'].dt.year
    df['met_year'] = df['year']
    df.loc[df['month'] == 12, 'met_year'] += 1 
    
    season_map = {12: 'Winter', 1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring',
                  6: 'Summer', 7: 'Summer', 8: 'Summer', 9: 'Autumn', 10: 'Autumn', 11: 'Autumn'}
    df['season'] = df['month'].map(season_map)
    
    agg_dict = {var: 'mean' for var in MEAN_VARS if var in available_vars}
    agg_dict.update({var: 'sum' for var in SUM_VARS if var in available_vars})
    
    seasonal_df = df.groupby(['met_year', 'season', 'latitude', 'longitude']).agg(agg_dict).reset_index()
    pivot_df = seasonal_df.pivot_table(index=['met_year', 'latitude', 'longitude'], columns='season', values=available_vars)
    pivot_df.columns = [f"{var}_{season}" for var, season in pivot_df.columns]
    pivot_df = pivot_df.reset_index().dropna()
    pivot_df['target'] = (pivot_df['met_year'] >= 1990).astype(int)
    return pivot_df

# --- 2. TUNING & STATS FUNCTIONS ---
def calculate_classification_z_score(y_true, y_pred, model_name):
    accuracy, n_samples, p_null = accuracy_score(y_true, y_pred), len(y_true), 0.50 
    standard_error = np.sqrt((p_null * (1 - p_null)) / n_samples)
    z_score = (accuracy - p_null) / standard_error
    p_value = stats.norm.sf(abs(z_score)) * 2 
    print(f"\n--- Z-Score for {model_name} ---")
    print(f"Accuracy: {accuracy:.2%} | Z-Score: {z_score:.2f} | P-Value: {p_value:.4e}")
    print("Conclusion: SIGNIFICANT (>95% confidence)." if z_score > 1.96 else "Conclusion: NOT SIGNIFICANT.")

def sequential_tuning(X_train, X_test, y_train, y_test, model_type):
    print(f"\n--- Tuning {model_type} ---")
    best_params = {'random_state': 42} if model_type != 'KNN' else {}
    
    if model_type == 'Random Forest':
        base_model = RandomForestClassifier
        best_params['n_jobs'] = -1
        tuning_sequence = [('n_estimators', [10, 50, 100]), ('max_depth', [5, 10, None]), ('min_samples_split', [2, 5, 10])]
    elif model_type == 'Boosted Decision Tree':
        base_model = HistGradientBoostingClassifier 
        tuning_sequence = [('max_iter', [50, 100, 150]), ('learning_rate', [0.05, 0.1, 0.2]), ('max_depth', [3, 5, 7])]
    else:
        base_model = KNeighborsClassifier
        best_params.update({'algorithm': 'kd_tree', 'n_jobs': 1})
        tuning_sequence = [('n_neighbors', [3, 5, 9, 15]), ('weights', ['uniform', 'distance'])]

    for param_name, param_values in tuning_sequence:
        best_score, best_val = 0, None
        for val in param_values:
            current_params = best_params.copy()
            current_params[param_name] = val
            model = base_model(**current_params)
            model.fit(X_train, y_train)
            score = accuracy_score(y_test, model.predict(X_test))
            if score > best_score: best_score, best_val = score, val
        best_params[param_name] = best_val
        print(f"Locked in {param_name} = {best_val} (Acc: {best_score:.2%})")
        
    final_model = base_model(**best_params)
    final_model.fit(X_train, y_train)
    return final_model

# --- EXECUTE DATA LOAD & LEAK-PROOF SPLIT ---
df_seasonal = load_and_prep_seasonal_data().sample(frac=0.25, random_state=42)

# Separate features, target, and the 'groups' (the year)
feature_cols = [c for c in df_seasonal.columns if c not in ['met_year', 'latitude', 'longitude', 'target']]
X = df_seasonal[feature_cols]
y = df_seasonal['target']
groups = df_seasonal['met_year']

# Generate the 75/25 split based on whole Years
gss = GroupShuffleSplit(n_splits=1, test_size=0.25, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups))

X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

# Prove the split worked correctly:
train_years = df_seasonal.iloc[train_idx]['met_year'].unique()
test_years = df_seasonal.iloc[test_idx]['met_year'].unique()
overlap = set(train_years).intersection(set(test_years))

print("\n--- Data Leakage Check ---")
print(f"Years in Train set: {len(train_years)}")
print(f"Years in Test set : {len(test_years)}")
print(f"Years in BOTH     : {len(overlap)} (Check this is 0)")
print("\nData loaded and ready for modeling!")
```

```python
# RandomForest Optimisation
rf_model = sequential_tuning(X_train, X_test, y_train, y_test, 'Random Forest')
calculate_classification_z_score(y_test, rf_model.predict(X_test), 'Random Forest')
```

```python
# Boosted Decision Tree Optimisation
bdt_model = sequential_tuning(X_train, X_test, y_train, y_test, 'Boosted Decision Tree')
calculate_classification_z_score(y_test, bdt_model.predict(X_test), 'Boosted Decision Tree')
```

```python
# kNN Optimisation
knn_model = sequential_tuning(X_train, X_test, y_train, y_test, 'KNN')
calculate_classification_z_score(y_test, knn_model.predict(X_test), 'KNN')
```

```python
# Plots
def plot_probability_densities(models_dict, X_test, y_test):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    for ax, (name, model) in zip(axes, models_dict.items()):
        y_prob = model.predict_proba(X_test)[:, 1]
        sns.histplot(y_prob[y_test == 0], color='blue', label='Actual Pre-1990', kde=True, stat="density", element="step", alpha=0.4, ax=ax)
        sns.histplot(y_prob[y_test == 1], color='red', label='Actual Post-1990', kde=True, stat="density", element="step", alpha=0.4, ax=ax)
        ax.axvline(0.5, color='black', linestyle='--', label='Decision Boundary')
        ax.set_title(f"{name} Confidence")
        ax.set_xlabel("Predicted Probability (Post-1990)")
        if ax == axes[0]: ax.set_ylabel("Density")
        ax.legend(loc='upper center', fontsize=9)
    plt.tight_layout()
    plt.show()

def plot_roc_curves(models_dict, X_test, y_test):
    plt.figure(figsize=(8, 6))
    for name, model in models_dict.items():
        y_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{name} (AUC = {roc_auc:.2f})')
        
    plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--', label='Random Guess')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves')
    plt.legend(loc="lower right")
    plt.grid(alpha=0.3)
    plt.show()

# Combine models into a dictionary for plotting
optimized_models = {
    'Random Forest': rf_model, 
    'Boosted Decision Tree': bdt_model, 
    'KNN': knn_model
}

# 1. Plot Densities & ROC Curves
plot_probability_densities(optimized_models, X_test, y_test)
plot_roc_curves(optimized_models, X_test, y_test)

# 2. Plot Feature Importances (Only for Tree models, as KNN doesn't support this)
tree_models = {'Random Forest': rf_model}
# Note: HistGradientBoosting does not expose .feature_importances_ natively like standard GBM does
fig, ax = plt.subplots(figsize=(10, 6))
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1][:15] # Top 15
sns.barplot(x=importances[indices], y=np.array(feature_cols)[indices], ax=ax, palette="viridis")
ax.set_title("Top 15 Feature Importances: Random Forest")
ax.set_xlabel("Relative Importance")
plt.tight_layout()
plt.show()
```
