# UK Climate Analysis: Pre-1990 vs. Post-1990 Classifcation
This notebook performs a supervised machine learning analysis to investigate evidence of climate change in the UK using ERA5 Reanalysis data.

## Objective
To determine if a Random Forest Classifier can accurately distinguish between historical climate conditions (Pre-1990) and modern climate conditions (Post-1990) based on physical meteorological variables.

## Methodology
Data Preprocessing: Loading ERA5 NetCDF data, converting units (Kelvin to Celsius, Pa to hPa), and deriving calculated variables (e.g., Wind Speed).

Feature Selection: Analysis of 10 key climate variables, including 2m Temperature, Total Precipitation, Soil Temperature, and Snowfall.

Classification: Training a Random Forest model on a 75/25 Train-Test split to classify individual grid cells/months into the two time periods.

Evaluation: Assessing model performance using Accuracy scores, Confusion Matrices, and ROC Curves.

## Key Visualizations
Correlation Matrix: To identify relationships between climate variables.

ROC Curves: To visualize the diagnostic ability of the classifier.

Feature Importance Ranking: To quantify which variables are the strongest indicators of the climate shift.

Decision Boundary & Probability Distributions: To visualize the statistical separation between the two climate eras.

```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc
from sklearn.tree import DecisionTreeClassifier

# ==========================================
# 1. SETUP
# ==========================================
# Update this path to your actual ERA5 file location
FILE_PATH = r'C:\Users\James\Downloads\era5new.nc'

# Variables selected for the report
VAR_LIST = [
    't2m',          # 2m Temperature
    'tp',           # Total Precipitation
    'stl2',         # Soil Temp Level 2
    'wind_speed',   # 10m Wind Speed (derived)
    'd2m',          # Dewpoint Temperature
    'e',            # Total Evaporation
    'ro',           # Runoff
    'sf',           # Snowfall
    'swvl1',        # Volumetric Soil Water Layer 1
    'sp'            # Surface Pressure
]
```

```python
# ==========================================
# 2. DATA PROCESSING FUNCTION
# ==========================================
def load_and_prep_data():
    print("--- Loading Dataset ---")
    ds = xr.open_dataset(FILE_PATH)
    
    # Handle naming differences in recent ERA5 downloads
    if 'valid_time' in ds.coords:
        ds = ds.rename({'valid_time': 'time'})
    
    # Derive Wind Speed if U/V components are available
    # Checks for both standard naming conventions
    u_var = ds['u10'] if 'u10' in ds else (ds['10u'] if '10u' in ds else None)
    v_var = ds['v10'] if 'v10' in ds else (ds['10v'] if '10v' in ds else None)
    
    if u_var is not None and v_var is not None:
        ds['wind_speed'] = np.sqrt(u_var**2 + v_var**2)
        print("   -> Calculated Wind Speed magnitude")
    
    # Unit adjustments for consistency
    # Convert Pressure to hPa
    if 'sp' in ds: ds['sp'] = ds['sp'] / 100.0
    
    # Convert Kelvin to Celsius where applicable
    for var in ['t2m', 'stl2', 'd2m']:
        if var in ds and ds[var].mean() > 200:
            ds[var] = ds[var] - 273.15
            
    print("--- Constructing DataFrame ---")
    
    # Filter dataset to only the chosen variables
    available_vars = list(set(VAR_LIST).intersection(ds.data_vars))
    ds_subset = ds[available_vars]
    
    # Convert xarray dataset to pandas DataFrame
    # Resulting index is MultiIndex (time, latitude, longitude)
    df = ds_subset.to_dataframe()
    
    # Remove rows with NaN values (e.g., ocean grid cells if using land variables)
    initial_rows = len(df)
    df = df.dropna()
    print(f"   -> Removed {initial_rows - len(df)} NaN rows")
    
    # Reset index to access 'time' as a column
    df = df.reset_index()
    
    # Define Target Variable: 1 for Post-1990, 0 for Pre-1990
    df['target'] = (df['time'].dt.year >= 1990).astype(int)
    
    # Check dimensions as requested
    n_lat = ds.dims['latitude']
    n_lon = ds.dims['longitude']
    n_time = ds.dims['time']
    print("\n--- Dimension Verification ---")
    print(f"Grid: {n_time} time steps x {n_lat} lat x {n_lon} lon")
    print(f"Total DataFrame Rows: {len(df)}")
    
    return df
```

```python
# ==========================================
# 3. ANALYSIS & MODELING FUNCTION
# ==========================================
def run_analysis(df):
    # Set up Features (X) and Target (y)
    feature_cols = [c for c in df.columns if c in VAR_LIST]
    X = df[feature_cols]
    y = df['target']
    
    # --- Correlation Analysis ---
    print("\n--- Generating Correlation Matrix ---")
    plt.figure(figsize=(10, 8))
    corr = X.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title("Correlation Matrix")
    plt.tight_layout()
    plt.show()

    # --- Model Training ---
    print("--- Training Random Forest (75:25 Split) ---")
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    
    # Initialize Random Forest
    rf = RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42, n_jobs=-1)
    rf.fit(X_train, y_train)
    
    # Generate Predictions
    y_pred_train = rf.predict(X_train)
    y_pred_test = rf.predict(X_test)
    y_prob_train = rf.predict_proba(X_train)[:, 1]
    y_prob_test = rf.predict_proba(X_test)[:, 1]
    
    # --- Performance Metrics ---
    print("\n--- Model Performance ---")
    print(f"Training Accuracy: {accuracy_score(y_train, y_pred_train):.2%}")
    print(f"Testing Accuracy:  {accuracy_score(y_test, y_pred_test):.2%}")
    print("\nClassification Report (Test Set):")
    print(classification_report(y_test, y_pred_test))

    # --- ROC Curves ---
    fpr_train, tpr_train, _ = roc_curve(y_train, y_prob_train)
    fpr_test, tpr_test, _ = roc_curve(y_test, y_prob_test)
    auc_train = auc(fpr_train, tpr_train)
    auc_test = auc(fpr_test, tpr_test)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr_train, tpr_train, label=f'Training ROC (AUC={auc_train:.2f})', linestyle='--')
    plt.plot(fpr_test, tpr_test, label=f'Testing ROC (AUC={auc_test:.2f})', linewidth=2)
    plt.plot([0, 1], [0, 1], 'k--', label='Chance')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve: Pre vs Post 1990')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    # --- Feature Importance ---
    importances = rf.feature_importances_
    indices = np.argsort(importances)[::-1]
    
    plt.figure(figsize=(10, 6))
    plt.title("Feature Importance Ranking")
    plt.bar(range(X.shape[1]), importances[indices], align="center", color='teal')
    plt.xticks(range(X.shape[1]), [X.columns[i] for i in indices], rotation=45, ha='right')
    plt.ylabel("Importance Score")
    plt.tight_layout()
    plt.show()

    # --- Decision Boundary Visualization ---
    print("\n--- Visualizing Decision Boundary ---")
    
    # Identify top feature
    top_idx = indices[0]
    top_feature_name = X.columns[top_idx]
    print(f"Most significant variable: {top_feature_name}")
    
    # Calculate optimal split using a simple decision stump
    tree = DecisionTreeClassifier(max_depth=1)
    tree.fit(X_train[[top_feature_name]], y_train)
    threshold = tree.tree_.threshold[0]
    print(f"Optimal threshold: {threshold:.2f}")

    # Plot distribution with boundary
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=top_feature_name, hue='target', kde=True, 
                 palette={0: 'blue', 1: 'red'}, element="step", stat="density", common_norm=False)
    
    plt.axvline(threshold, color='green', linestyle='--', linewidth=3, label=f'Boundary ({threshold:.2f})')
    
    plt.title(f"Distribution of {top_feature_name}\n(Pre-1990 vs Post-1990)")
    plt.legend(title='Period', labels=['Decision Boundary', 'Post-1990', 'Pre-1990'])
    plt.xlabel(top_feature_name)
    plt.grid(True, alpha=0.3)
    plt.show()
    
    # RETURN the test data so we can use it outside this function
    return y_test, y_prob_test
```

```python
# ==========================================
# 4. EXECUTION
# ==========================================

# 1. Load Data
df_clean = load_and_prep_data()

# 2. Run Main Analysis and capture the returned variables
y_test, y_prob_test = run_analysis(df_clean)

# 3. Plot the Probability Distribution
plt.figure(figsize=(10, 6))

# Plot probabilities for actual Pre-1990 years
sns.histplot(y_prob_test[y_test == 0], color='blue', label='Actual Pre-1990', 
             kde=True, stat="density", element="step", alpha=0.4)

# Plot probabilities for actual Post-1990 years
sns.histplot(y_prob_test[y_test == 1], color='red', label='Actual Post-1990', 
             kde=True, stat="density", element="step", alpha=0.4)

# Add decision line at 0.5
plt.axvline(0.5, color='black', linestyle='--', linewidth=2, label='Decision Boundary (0.5)')

plt.title("Model Prediction Confidence")
plt.xlabel("Predicted Probability (0 = Pre-1990, 1 = Post-1990)")
plt.ylabel("Density")
plt.legend(loc='upper center')
plt.grid(True, alpha=0.3)
plt.show()
```
