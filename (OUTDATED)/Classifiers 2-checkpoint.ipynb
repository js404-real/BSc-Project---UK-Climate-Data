```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
# from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score

# ==========================================
# 1. SETUP & CONFIGURATION
# ==========================================
FILE_PATH = r'C:\Users\James\Downloads\era5new.nc'

MEAN_VARS = ['t2m', 'stl2', 'wind_speed', 'd2m', 'swvl1', 'sp']
SUM_VARS = ['tp', 'e', 'ro', 'sf']
VAR_LIST = MEAN_VARS + SUM_VARS

```

```python
# ==========================================
# 2. DATA PROCESSING
# ==========================================
def load_and_prep_seasonal_data(file_path=FILE_PATH):
    """
    Loads ERA5 NetCDF data, computes derived variables, and aggregates
    data into seasonal metrics (Spring, Summer, Autumn, Winter).
    
    Returns:
        pd.DataFrame: A flattened DataFrame with seasonal features per grid cell and year.
    """
    print("Loading dataset...")
    ds = xr.open_dataset(file_path)
    
    if 'valid_time' in ds.coords:
        ds = ds.rename({'valid_time': 'time'})
        
    u_var = ds['u10'] if 'u10' in ds else (ds['10u'] if '10u' in ds else None)
    v_var = ds['v10'] if 'v10' in ds else (ds['10v'] if '10v' in ds else None)
    
    if u_var is not None and v_var is not None:
        ds['wind_speed'] = np.sqrt(u_var**2 + v_var**2)
        
    if 'sp' in ds: 
        ds['sp'] = ds['sp'] / 100.0
        
    for var in ['t2m', 'stl2', 'd2m']:
        if var in ds and ds[var].mean() > 200:
            ds[var] = ds[var] - 273.15
            
    available_vars = list(set(VAR_LIST).intersection(ds.data_vars))
    df = ds[available_vars].to_dataframe().dropna().reset_index()
    
    print("Applying seasonal aggregation...")
    df['month'] = df['time'].dt.month
    df['year'] = df['time'].dt.year
    
    # Shift December to the subsequent meteorological year
    df['met_year'] = df['year']
    df.loc[df['month'] == 12, 'met_year'] += 1 
    
    season_map = {
        12: 'Winter', 1: 'Winter', 2: 'Winter',
        3: 'Spring', 4: 'Spring', 5: 'Spring',
        6: 'Summer', 7: 'Summer', 8: 'Summer',
        9: 'Autumn', 10: 'Autumn', 11: 'Autumn'
    }
    df['season'] = df['month'].map(season_map)
    
    agg_dict = {var: 'mean' for var in MEAN_VARS if var in available_vars}
    agg_dict.update({var: 'sum' for var in SUM_VARS if var in available_vars})
    
    seasonal_df = df.groupby(['met_year', 'season', 'latitude', 'longitude']).agg(agg_dict).reset_index()
    
    pivot_df = seasonal_df.pivot_table(
        index=['met_year', 'latitude', 'longitude'],
        columns='season',
        values=available_vars
    )
    
    pivot_df.columns = [f"{var}_{season}" for var, season in pivot_df.columns]
    pivot_df = pivot_df.reset_index().dropna()
    
    pivot_df['target'] = (pivot_df['met_year'] >= 1990).astype(int)
    
    print(f"Final structured DataFrame shape: {pivot_df.shape}")
    
    return pivot_df

```

```python
# ==========================================
# 3. MODELING & EVALUATION
# ==========================================
def run_advanced_models(df):
    """
    Trains baseline classifiers (Random Forest, Gradient Boosting) 
    and plots their classification probability distributions.
    """
    feature_cols = [c for c in df.columns if c not in ['met_year', 'latitude', 'longitude', 'target']]
    X = df[feature_cols]
    y = df['target']
    
    print(f"Training models on {len(feature_cols)} features...")
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    
    results = {}

    # --- 1. Random Forest ---
    rf_model = RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42, n_jobs=-1)
    rf_model.fit(X_train, y_train)
    results['Random Forest'] = accuracy_score(y_test, rf_model.predict(X_test))
    
    # --- 2. Boosted Decision Tree (GBM) ---
    gbm_model = GradientBoostingClassifier(n_estimators=50, learning_rate=0.1, max_depth=3, random_state=42)
    gbm_model.fit(X_train, y_train)
    results['Boosted Decision Tree'] = accuracy_score(y_test, gbm_model.predict(X_test))

    # --- 3. K-Nearest Neighbors (KNN) ---
    # Temporarily commented out due to environment threadpoolctl bug
    # knn_model = KNeighborsClassifier(n_neighbors=5, weights='uniform')
    # knn_model.fit(X_train, y_train)
    # results['KNN'] = accuracy_score(y_test, knn_model.predict(X_test))

    print("\nModel Accuracy Comparison:")
    print("-" * 40)
    for model_name, acc in results.items():
        print(f"{model_name:<25}: {acc:.2%}")

    print("\nGenerating probability distributions...")
    
    trained_models = {
        'Random Forest': rf_model,
        'Boosted Decision Tree': gbm_model,
        # 'KNN': knn_model
    }
    
    # Changed from 1x3 to 1x2 to account for the removed KNN model
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    for ax, (model_name, model) in zip(axes, trained_models.items()):
        y_prob = model.predict_proba(X_test)[:, 1]
        
        sns.histplot(y_prob[y_test == 0], color='blue', label='Actual Pre-1990', 
                     kde=True, stat="density", element="step", alpha=0.4, ax=ax)
        
        sns.histplot(y_prob[y_test == 1], color='red', label='Actual Post-1990', 
                     kde=True, stat="density", element="step", alpha=0.4, ax=ax)
        
        ax.axvline(0.5, color='black', linestyle='--', linewidth=2, label='Decision Boundary (0.5)')
        
        ax.set_title(f"{model_name} Confidence")
        ax.set_xlabel("Predicted Probability (Post-1990)")
        
        if ax == axes[0]:
            ax.set_ylabel("Density")
        else:
            ax.set_ylabel("") 
            
        ax.legend(loc='upper center', fontsize=9)
        ax.grid(True, alpha=0.3)
        
    plt.tight_layout()
    plt.show()
        
    return rf_model, gbm_model

```

```python
# ==========================================
# 4. EXECUTION
# ==========================================
if __name__ == "__main__":
    df_seasonal = load_and_prep_seasonal_data()
    # Utilizing a 10% sample to manage computation limits during testing
    models = run_advanced_models(df_seasonal.sample(frac=0.1, random_state=42))
```
