# UK Climate Data: Various Plots

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np

# Use raw string (r'...') to avoid Windows path errors
FILE_PATH = r'C:\Users\James\Downloads\era5new.nc' 

# Configuration for all 10 variables
# keys must match the variable names in the processed dataset
VAR_CONFIG = {
    't2m':        {'type': 'avg', 'label': '2m Temperature (째C)', 'cmap': 'RdBu_r'},
    'tp':         {'type': 'sum', 'label': 'Total Precipitation (mm)', 'cmap': 'Blues'},
    'stl2':       {'type': 'avg', 'label': 'Soil Temp Level 2 (째C)', 'cmap': 'inferno'},
    'wind_speed': {'type': 'avg', 'label': '10m Wind Speed (m/s)', 'cmap': 'YlGnBu'},
    'd2m':        {'type': 'avg', 'label': 'Dewpoint Temperature (째C)', 'cmap': 'Greens'},
    'e':          {'type': 'sum', 'label': 'Total Evaporation (mm)', 'cmap': 'BrBG'},
    'ro':         {'type': 'sum', 'label': 'Runoff (mm)', 'cmap': 'PuBu'},
    'sf':         {'type': 'sum', 'label': 'Snowfall (mm water eq)', 'cmap': 'PuBuGn'},
    'swvl1':      {'type': 'avg', 'label': 'Volumetric Soil Water (m3/m3)', 'cmap': 'gist_earth_r'},
    'sp':         {'type': 'avg', 'label': 'Surface Pressure (hPa)', 'cmap': 'RdGy_r'},
}
```

```python
def preprocess_dataset(ds):
    """
    Clean up time dimensions and calculate derived variables.
    """
    # 1. Fix Time Name (valid_time -> time)
    if 'valid_time' in ds.coords:
        ds = ds.rename({'valid_time': 'time'})
        print("Renamed 'valid_time' to 'time'.")
    
    # 2. Safe Wind Speed Calculation
    # We check if the keys exist in the file first, rather than using 'or' on the data
    u_var = None
    if 'u10' in ds:
        u_var = ds['u10']
    elif '10u' in ds:
        u_var = ds['10u']
        
    v_var = None
    if 'v10' in ds:
        v_var = ds['v10']
    elif '10v' in ds:
        v_var = ds['10v']
    
    # Only calculate if we found BOTH components
    if u_var is not None and v_var is not None:
        ds['wind_speed'] = np.sqrt(u_var**2 + v_var**2)
        print("Calculated Wind Speed from components.")
    else:
        print("Could not find U/V wind components. Skipping wind speed calc.")
    
    # 3. Pressure Conversion (Pa -> hPa)
    if 'sp' in ds:
        ds['sp'] = ds['sp'] / 100.0
        
    return ds

def get_yearly_data(ds, var_name, config):
    """
    Aggregates data to yearly averages or sums.
    """
    da = ds[var_name]
    
    # --- TYPE: AVERAGE ---
    if config['type'] == 'avg':
        # Convert Kelvin to Celsius for Temperature variables (check mean > 200)
        # We only do this for variables that look like Kelvin
        if da.mean() > 200 and 'Pa' not in config['label']: 
            da = da - 273.15
        da_yearly = da.groupby('time.year').mean(dim='time')

    # --- TYPE: SUM ---
    elif config['type'] == 'sum':
        # Multiply rate by days in month
        days_in_month = da.time.dt.days_in_month
        da_accumulated = da * days_in_month
        
        # Convert meters to mm for water variables
        if 'mm' in config['label']:
             da_accumulated = da_accumulated * 1000
        
        da_yearly = da_accumulated.groupby('time.year').sum(dim='time')
        
    return da_yearly

def plot_variable(ds, var_key):
    """
    Generates Map, Time Series, and Histograms for a single variable.
    """
    if var_key not in ds:
        print(f"Error: Variable '{var_key}' not found in dataset.")
        return

    config = VAR_CONFIG[var_key]
    print(f"--- Plotting {config['label']} ---")
    
    # Get yearly data
    da_yearly = get_yearly_data(ds, var_key, config)

    # 1. MAP
    fig = plt.figure(figsize=(10, 6))
    ax = plt.axes(projection=ccrs.PlateCarree())
    da_yearly.mean(dim='year').plot(ax=ax, cmap=config['cmap'], 
                                    cbar_kwargs={'label': config['label']})
    ax.coastlines()
    ax.set_title(f"Climatological Mean: {config['label']}")
    plt.show()

    # 2. TIME SERIES
    ts_mean = da_yearly.mean(dim=['latitude', 'longitude'])
    ts_std = da_yearly.std(dim=['latitude', 'longitude'])
    
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(ts_mean.year, ts_mean, '-o', color='black', lw=1, ms=3, label='UK Mean')
    ax.fill_between(ts_mean.year, ts_mean - ts_std, ts_mean + ts_std, 
                    color='gray', alpha=0.3, label='Spatial Variability (1 Sigma)')
    ax.set_title(f"Annual Trend: {config['label']}")
    ax.set_ylabel(config['label'])
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.legend()
    plt.show()

    # 3. HISTOGRAMS
    pre_90 = da_yearly.sel(year=slice(None, 1989)).mean(dim=['latitude', 'longitude'])
    post_90 = da_yearly.sel(year=slice(1990, None)).mean(dim=['latitude', 'longitude'])
    
    fig, ax = plt.subplots(figsize=(8, 5))
    
    # Calculate shared bins
    data_min = min(pre_90.min(), post_90.min())
    data_max = max(pre_90.max(), post_90.max())
    bins = np.linspace(float(data_min), float(data_max), 20)
    
    ax.hist(pre_90, bins=bins, alpha=0.5, color='blue', label='Pre-1990', density=True)
    ax.hist(post_90, bins=bins, alpha=0.5, color='red', label='Post-1990', density=True)
    ax.set_xlabel(config['label'])
    ax.set_title(f"Distribution Shift: {config['label']}")
    ax.legend()
    plt
```

```python
# Open and process the dataset
ds_raw = xr.open_dataset(FILE_PATH)
ds = preprocess_dataset(ds_raw)
print("Dataset loaded and processed successfully.")
print("Available variables:", list(ds.data_vars))
```
